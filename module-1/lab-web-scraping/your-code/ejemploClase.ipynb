```python
url = 'https://es.wikipedia.org/wiki/Videoconsola'
```

```python
resp = requests.get(url)
sopa_v = BeautifulSoup(resp.content,'html.parser')
sopa_v
```

```python
tables = sopa_v.find_all('table',{'class':'wikitable sortable'})
```

```python
tables
```

```python
españa = tables[-1]
```

```python
cabe_esp = españa.findAll('th')
cabe_esp
```

```python
cabe_txt_v = [i.text.replace('\n','') for i in cabe_esp]
cabe_txt_v
```

```python
rows = españa.find_all('tr')
rows_clean = [i.text.strip().split('\n') for i in rows]
rows_clean
```

```python
rows_clean.pop(0)
```

```python
final = []
for i in rows_clean:
    tmp = []
    for e in i:
        if e != '':
            tmp.append(e)
    final.append(tmp)
    
final
```

```python
df_final = pd.DataFrame(final,columns = cabe_txt_v)
df_final
```

```python
clean_uni = lambda x: re.match('(\d+ millones| \d+millon)',x).group()[0]
```

```python
df_final['Unidades vendidas'] = df_final['Unidades vendidas'].apply(clean_uni)
```

```python
df_final['Unidades vendidas'] = df_final['Unidades vendidas'].replace(r'\[.*?\]|\(.*?\)','', regex = True)
```

```python
df_final['Unidades vendidas'] = df_final['Unidades vendidas'].apply(lambda x: re.findall('\d+',x))
```

```python
df_final
```

```python
df_final['Lanzamiento'] = df_final['Lanzamiento'].replace(r'\[.*?\]|\(.*?\)','', regex = True)
```

```python
df_final
```
