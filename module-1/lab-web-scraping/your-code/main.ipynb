#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
req = requests.get(url)
req_text = req.text
sopa = BeautifulSoup(req.content, 'html.parser')
````

````python
#clases = sopa.find_all(attrs = {"class" : "Box-row"})
clases = sopa.find_all(attrs = {"class" : "h3 lh-condensed"})
clases[0]
````

````python
def obtener_repo(clase):
    for link in clase.find_all('a', href=True):
        return link['href']
    
repos = [obtener_repo(clase) for clase in clases]
repos
````

#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
req = requests.get(url)
req_text = req.text
sopa = BeautifulSoup(req.content, 'html.parser')
````

````python
def obtener_imagenes(clase):
    imagenes = []
    for img in clase.find_all('img'):
        imagenes.append(img['src'])
    return imagenes
    
imagenes = obtener_imagenes(sopa)
imagenes
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python'
req = requests.get(url)
req_text = req.text
sopa = BeautifulSoup(req.content, 'html.parser')
````

````python
def obtener_links(clase):
    links = []
    for link in clase.find_all('a', href=True):
        if "https" in str(link['href']):
            links.append(link['href'])
    return links
links = obtener_links(sopa)
links
````

#### Number of Titles that have changed in the United States Code since its last release point 

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'
req = requests.get(url)
req_text = req.text
sopa = BeautifulSoup(req.content, 'html.parser').find_all(attrs = {"class" : "uscitemlist"})
sopa
````

````python
def obtener_titulos(clase):
    titulos = []
    for titulo in clase.find_all(attrs={"class":"usctitlechanged"}):
        titulos.append(titulo.text.strip())
    return titulos

titulos = obtener_titulos(sopa[0])
titulos
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
req = requests.get(url)
req_text = req.text
sopa = BeautifulSoup(req.content, 'html.parser').find_all('tbody',attrs = {"id" : "tbody"})
sopa
````

````python
filas = sopa[0].find_all('tr')
len(filas)
filas[0]
````

````python
earthquakes = []
for i in range(13):
    date = (filas[i].find_all('a', href=True)[0]).text[:10]
    time = (filas[i].find_all('a', href=True)[0]).text[10:-2]
    latitude = (filas[i].find_all('td', attrs = {"class" : "tabev1"})[0]).text
    longitude = (filas[i].find_all('td', attrs = {"class" : "tabev1"})[1]).text
    region = (filas[i].find_all('td', attrs = {"class":'tb_region'})[0]).text
    print(date,time,latitude,longitude,region)
    earthquakes.append([date,time,latitude,longitude,region])
    
pd.DataFrame(earthquakes, columns= ['date', 'time', 'lat', 'lon', 'region'])
````
