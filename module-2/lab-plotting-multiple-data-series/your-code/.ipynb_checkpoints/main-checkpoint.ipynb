# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
filtered_1 = data[["ItemType","RetailSales", "RetailTransfers", "WarehouseSales"]]
```

```python
filtered_1 = filtered_1.pivot_table(filtered_1, index = "ItemType", aggfunc = sum)
filtered_1.plot(kind="bar")
plt.show()
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
filtered_2 = data[["Supplier"]]
filtered_2["TotalSales"] = data["RetailSales"]+data["RetailTransfers"]+data["WarehouseSales"]
filtered_2 = filtered_2.pivot_table(filtered_2, index= "Supplier")
filtered_2 = filtered_2.sort_values(by= ["TotalSales"], ascending = False)
filtered_2= filtered_2[:10]
filtered_2
```

```python
filtered_2.plot(kind="barh")
plt.show()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
filtered_3 = data[["Month","RetailSales", "RetailTransfers", "WarehouseSales"]]
filtered_3 = filtered_3.pivot_table(index= "Month", values = ["RetailSales", "RetailTransfers", "WarehouseSales"], aggfunc= "mean")
filtered_3
```

```python
filtered_3.plot.line()
```

## 4. Plot the same information as above but as a bar chart.

```python
filtered_3.plot.bar()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
filtered_4 = data[["ItemType","Year","Month","RetailSales"]]
filtered_4 = filtered_4.pivot_table(index= ["Year","Month"], values = ["RetailSales"], columns = "ItemType", aggfunc="sum")
filtered_4
```

```python
filtered_4.plot.line()
```

## 6. Plot the same information as above but as a bar chart.

```python
filtered_4.plot.bar()
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
#Mayumi, aqui lo que hago es crear un mapa de colores aleatorios y los cambio por los valores de item type
filtered_5 = data[["ItemType","RetailSales","RetailTransfers"]]
mapa = {}
uniques = list(filtered_5["ItemType"].unique())
colors = ["#973a43","#3c73fe","#fd4b94","#a4880c", "#1f8489", "#d13523", "#e4ac7c", "#b3eccc", "#b2fe2d"]
color_map = {uniques[i]: colors[i] for i in range(len(uniques))}
```

```python
color_map["BEER"]
```

```python
filtered_5["ItemType"] = filtered_5["ItemType"].apply(lambda x : color_map[x])
filtered_5.head()
```

```python
color_map
```

```python
#Lo anterior lo hice para poder usar aca los colores en el argumento 'c'
filtered_5.plot.scatter(x="RetailSales", y="RetailTransfers", c = filtered_5.ItemType)
```
